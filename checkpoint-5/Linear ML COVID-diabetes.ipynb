```
paises = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/frodriguez@insulet.com/pai_ses-2.csv")
```

```
from pyspark.sql.functions import col

paises = paises.withColumn("total_cases_per_million", col("total_cases_per_million").cast("double"))
paises = paises.withColumn("total_deaths_per_million", col("total_deaths_per_million").cast("double"))
paises = paises.withColumn("population_density", col("population_density").cast("double"))
paises = paises.withColumn("cardiovasc_death_rate", col("cardiovasc_death_rate").cast("double"))
paises = paises.withColumn("diabetes_prevalence", col("diabetes_prevalence").cast("double"))
paises = paises.withColumn("life_expectancy", col("life_expectancy").cast("double"))
paises = paises.withColumn("hdi", col("hdi").cast("double"))


# casting strings for doubles

paises
```

```
paises = paises.na.drop()
```

## Train/Test Split

![](https://files.training.databricks.com/images/301/TrainTestSplit.png)

**Question**: Why is it necessary to set a seed? What happens if I change my cluster configuration?

```
trainDF, testDF = paises.randomSplit([.8, .2], seed=42)
print(trainDF.cache().count())
```

Let's change the # of partitions (to simulate a different cluster configuration), and see if we get the same number of data points in our training set.

```
trainRepartitionDF, testRepartitionDF = (paises
                                         .repartition(24)
                                         .randomSplit([.8, .2], seed=42))

print(trainRepartitionDF.count())

# we just simulated a different cluster
# despite same seed 42, we have 146 rows instead of 147 (previous cell) 
```

```
from pyspark.ml.regression import LinearRegression
```

```
display(trainDF.select("total_cases_per_million", "diabetes_prevalence"))
```

```
from pyspark.ml.feature import VectorAssembler

# we want to output the values of chosen columns into a single column, as a vector, and we will call it features
vecAssembler = VectorAssembler(inputCols=["diabetes_prevalence"], outputCol="features")
# we could do vecAssembler = VectorAssembler(inputCols=["variable1", "variable2", "variablen"], outputCol="features")

# transform takes a df and changes columns, or appends columns
# combining column values into a single column
vecTrainDF = vecAssembler.transform(trainDF)
```

```
display(vecTrainDF.select("total_cases_per_million", "features"))
# it looks funny, but the last element in features represents the vector
# so, it will be the diabetes prevalence
```

```
lr = LinearRegression(featuresCol="features", labelCol="total_cases_per_million")
lrModel = lr.fit(vecTrainDF)
```

## Inspect the model

```
m = lrModel.coefficients[0]
b = lrModel.intercept

print(f"The formula for the linear regression line is y = {m:.2f}x + {b:.2f}")

# En este caso nuestra x es el valor de la prevalencia de diabetes. 
# Suponiendo que un pais tuviera prevalencia de diabetes = 0, esperariamos 41,471.94 casos por millon
# Si un pais tuviera prevalencia de diabetes = 1, esperariamos 44,195.91 casos por millon
```

## Apply model to test set

```
vecTestDF = vecAssembler.transform(testDF)

predDF = lrModel.transform(vecTestDF)

predDF.select("diabetes_prevalence", "features", "total_cases_per_million", "prediction").show()
```

## Evaluate Model

Let's see how our linear regression model with just one variable does. Does it beat our baseline model?

```
from pyspark.ml.evaluation import RegressionEvaluator

regressionEvaluator = RegressionEvaluator(predictionCol="prediction", labelCol="total_cases_per_million", metricName="rmse")

rmse = regressionEvaluator.evaluate(predDF)
print(f"RMSE is {rmse}")

r2 = regressionEvaluator.setMetricName("r2").evaluate(predDF)
print(f"r2 is {r2}")
```

```
import mlflow
import mlflow.spark #spark "flavor"
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
```

```
vecAssembler2 = VectorAssembler(inputCols=["population_density", "cardiovasc_death_rate", "diabetes_prevalence", "life_expectancy", "hdi"], outputCol="features")
# transform takes a df and changes columns, or appends columns
# combining column values into a single column
vecTrainDF2 = vecAssembler2.transform(trainDF)
display(vecTrainDF2.select("total_cases_per_million", "features"))
```

```
lr2 = LinearRegression(featuresCol="features", labelCol="total_cases_per_million")
lrModel2 = lr2.fit(vecTrainDF2)
```

```
m1 = lrModel2.coefficients[0]
m2 = lrModel2.coefficients[1]
m3 = lrModel2.coefficients[2]
m4 = lrModel2.coefficients[3]
m5 = lrModel2.coefficients[4]
b = lrModel2.intercept


print(f"The formula for the linear regression line is y = {m1:.2f}x1 + {m2:.2f}x2 + {m3:.2f}x3 + {m4:.2f}x4 + {m5:.2f}x5 + {b:.2f}")
```

```
vecTestDF2 = vecAssembler2.transform(testDF)

predDF2 = lrModel2.transform(vecTestDF2)

predDF2.select("features", "total_cases_per_million", "prediction").show()
```

```
regressionEvaluator = RegressionEvaluator(predictionCol="prediction", labelCol="total_cases_per_million", metricName="rmse")

rmse2 = regressionEvaluator.evaluate(predDF2)
print(f"RMSE is {rmse2}")

r22 = regressionEvaluator.setMetricName("r2").evaluate(predDF2)
print(f"r2 is {r22}")
```
